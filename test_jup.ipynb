```python
if ppl ==  0:
    raise ValueError('wrong number of people')
```

```python

try:
    ppl = 6
    total = 10
    print(total/ppl)
except ZeroDivisionError:
    print("number of people cant be equal to zero")
except TypeError:
    print("ppl must be int")
except NameError:
    print("variable not defined")
except Exception:
    print("Big mistake")
else:
    print("we fulfill the else condition of try/exept block")
finally:
    print("Finishing exeption block")
```

```python
tips = 0.1
while True:
    try:
        total_bill = float(input("invoice amount: "))
        break
    except ValueError:
        print("Pls, enter correct value")

while True:
    try:
        guests = int(input("Number of guests: "))
        break
    except ValueError:
        print("Pls, enter an integer number of guests: ")

print(f"General tips {total_bill*tips:.2f}")
print(f"How much tips from each guest {(total_bill*tips)/guests:.2f}")
```

```python
class ValidateError(ValueError):
     pass

tips = 0.1
while True:
    try:
        total_bill = float(input("invoice amount: "))
        break
    except ValueError:
        print("Pls, enter correct value")

while True:
    try:
        guests = int(input("Number of guests: "))
        if guests <1:
            raise ValidateError("Less then 1 guests")
        break
    except ValidateError as err:
        print(f"Validation error: {err}")
    except ValueError:
        print("Pls, enter an integer number of guests: ")
    

print(f"General tips {total_bill*tips:.2f}")
print(f"How much tips from each guest {(total_bill*tips)/guests:.2f}")
```
